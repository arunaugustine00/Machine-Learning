```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 16, 6
```

```python
from sklearn import linear_model 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import r2_score
```

```python
dummy_data = {}
for i in range(200):
    dummy_data[i] = random.randint(i,i+20)
```

```python
data = pd.DataFrame.from_dict(dummy_data,orient='index')
```

```python
data.columns = ['univariat']
```

```python
plt.scatter(data.index,data.univariat)
plt.show()
```

```python
X = np.column_stack([np.ones(len(data), dtype=np.float32),data['univariat'].values])
y = data.index
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) 
```

```python
model_lr = linear_model.LinearRegression()
```

```python
model_lr.fit(X_train,y_train)
```

```python
y_predict = model_lr.predict(X_test)
```

```python
r2_score(y_test, y_predict)
```

```python
plt.scatter(X_test[:,1],y_test)
plt.scatter(X_test[:,1],y_predict)
plt.show()
```
